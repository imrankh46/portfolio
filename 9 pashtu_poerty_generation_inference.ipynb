```
import gradio as gr
import torch
path = 'Imran1/bloom_pg'
from transformers import pipeline
pipe = pipeline('text-generation',model=path, tokenizer=path,device=0)

```

```
import warnings
warnings.filterwarnings("ignore")



def generate_poetry(prompt):


    # Generate poetry
    output = pipe(
        prompt,
        top_p=0.9,
        top_k=200,
        temperature=0.7,
        do_sample=True,
        max_length=200
    )


    return output[0]['generated_text']

prompt = "ستا په راتلو شوم له ديدنه محروم"
p = f"""
Below is an instruction that describes a task. Write a response that appropriately completes the request.
### Instruction:
{prompt}
### Response:
"""

print(generate_poetry(p))
```
