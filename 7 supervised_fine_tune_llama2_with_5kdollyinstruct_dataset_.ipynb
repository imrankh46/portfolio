```python
import torch
from peft import AutoPeftModelForCausalLM
from transformers import AutoTokenizer

path= "Imran1/llama2-d"

# load base LLM model and tokenizer
model = AutoPeftModelForCausalLM.from_pretrained(
    path,
    low_cpu_mem_usage=True,
    torch_dtype=torch.float16,
    load_in_4bit=True,
)
tokenizer = AutoTokenizer.from_pretrained(path)
```

```python
from IPython.display import display, Markdown
import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')
def make_inference(instruction, context = None):
  if context:
    prompt = f"Below is an instruction that describes a task, paired with an input that provides further context.\n\n### Instruction: \n{instruction}\n\n### Input: \n{context}\n\n### Response: \n"
  else:
    prompt = f"Below is an instruction that describes a task. Write a response that appropriately completes the request.\n\n### Instruction: \n{instruction}\n\n### Response: \n"
  inputs = tokenizer(prompt, return_tensors="pt", return_token_type_ids=False).to("cuda:0")
  outputs = model.generate(**inputs, max_new_tokens=200,do_sample=True, top_p=0.9,temperature=0.9)
  display(Markdown((tokenizer.decode(outputs[0], skip_special_tokens=True))))
```

```python
make_inference("you are helpfull assistant for email writing.", "write email for me for machine learing engineer position at hugging face.")
```

```python
prompt = """
You are a talented story writer who can evoke vivid imagery and emotions through words.
"""
```

```python
make_inference(prompt, "Write a story of Smith and Sara, who meet at 25, fall in love but face challenges. Explore their failed love against the city's backdrop")
```
